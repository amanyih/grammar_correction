```python
import pickle
import matplotlib.pyplot as plt
from happytransformer import HappyTextToText, TTTrainArgs
```

```python
happy_tt = HappyTextToText("T5", "t5-base")
```

```python
eval_args = TTTrainArgs(
    batch_size=1,
    max_input_length=512,
    max_output_length=150
)

print("Evaluating the model BEFORE training...")
eval_results_before = happy_tt.eval("Dataset/eval.csv", args=eval_args)
print("Evaluation Loss (Before Training):", eval_results_before.loss)

with open("eval_results_before.txt", "w") as f:
    f.write("Evaluation Results Before Training:\n")
    f.write(str(eval_results_before))
```

 Train Model

```python
train_args = TTTrainArgs(
    batch_size=8,
    num_train_epochs=5,
    max_input_length=512,  
    max_output_length=150,
    eval_ratio=0.1
)

print("Training the model...")
training_losses = happy_tt.train("Dataset/train.csv", args=train_args)
```

Evaluate the model on the test set

```python
print("Evaluating the model AFTER training...")
eval_results_after = happy_tt.eval("Dataset/eval.csv", args=eval_args)
eval_loss_after = eval_results_after['eval_loss']
print("Evaluation Loss (After Training):", eval_loss_after)

with open("eval_results_after.txt", "w") as f:
    f.write("Evaluation Results After Training:\n")
    f.write(str(eval_results_after))
```

Plot Results

```python
epochs = list(range(1, len(training_losses) + 1))
plt.figure(figsize=(10, 6))
plt.plot(epochs, training_losses, label="Training Loss", marker='o')
plt.axhline(y=eval_loss_before, color='orange', linestyle='--', label=f"Evaluation Loss Before ({eval_loss_before:.4f})")
plt.axhline(y=eval_loss_after, color='green', linestyle='--', label=f"Evaluation Loss After ({eval_loss_after:.4f})")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Evaluation Losses")
plt.legend()
plt.grid(True)

plt.savefig("training_eval_comparison.png")
plt.show()
```

Final Results

```python
print("\nFinal Results:")
print(f"Evaluation Loss Before Training: {eval_loss_before:.4f}")
print(f"Evaluation Loss After Training: {eval_loss_after:.4f}")
```
